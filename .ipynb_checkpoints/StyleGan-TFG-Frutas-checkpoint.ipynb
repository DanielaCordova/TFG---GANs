## Red de mapeo $z$ → $w$

Tomamos vector ruido y lo mapea a un ruido intermedio. Hace que vector z este en un espacio mas desenredado, luego las funciones serán más fáciles de controlar adelante. La red de mapeo esta compuesta por 8 capas pero se implementa con una red neuronal de 3 capas. Se puede aumentar pero ahorras tiempo en el entrenamiento posterior con solo 3.



```python
class MapeoCapas(nn.Module):
    def __init__(self, z_dim, hidden_dim, w_dim):
        super().__init__()
        self.mapeo = nn.Sequential(
            nn.Linear(z_dim,hidden_dim),  #Aplica una transformación lineal a los datos entrantes: y=xA^T +b -> tranforma de z_dim a hidden_dim
            nn.ReLU(),
            nn.Linear(hidden_dim,hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, w_dim)
        )

    def forward(self, noise): #Dado un tensor de ruido inicial, devuelve el tensor de ruido intermedio.
        return self.mapeo(noise)
    
    def get_mapeo(self):
        return self.mapeo
```

## Inyección de ruido aleatoria
Ruido antes de cada bloque AdaIN. Se debe crear un vector de ruido del mismo tamaño que el mapa de características.
No es totalmente aleatorio si no que se inicializa con canal aleaotrio que luego se multiplica por los pesos aprendidos para cada canal en la imagen. 
Ejemplo: para una imagen tiene 512 canales con alto y ancho de (4 x 4). Primero se crea una matriz de ruido aleatoria (4 x 4) con un canal. Luego el modelo crea 512 valores, uno para cada canal. Luego, multiplica la matriz (4 x 4) por cada uno de estos valores. Esto crea un tensor "aleatorio" de 512 canales y (4 x 4) píxeles, las mismas dimensiones que la imagen. Finalmente, agrega este tensor de ruido a la imagen. Esto introduce ruido no correlacionado y está destinado a aumentar la diversidad en la imagen.


Se generan nuevos pesos iniciales para cada nueva capa, o generador, donde se usa esta clase. Dentro de una capa, cada vez que se llama a la inyección de ruido, da otro paso con el optimizador y los pesos que usa para cada canal se optimizan (es decir, se aprenden).

```python
class InyeccionDeRuido(nn.Module):
    def __init__(self, inChan): #inChan: num canales que la imagen tiene
        super().__init__()
        self.peso = nn.Parameter( # para que estos pesos se puedan optimizar
            torch.randn(inChan)[None, :, None, None] # Iniciar los pesos de los canales a partir de una distribución normal aleatoria. torch.randn((1,channels,1,1)) 
        )

    def forward(self, img): ## Para seguir adelante, dada imagen devuelve la misma pero con ruido agregado
        dimRuido = (img.shape[0], 1, img.shape[2], img.shape[3])  
        ruido = torch.randn(dimRuido, device=img.device) # Creacion del ruido
        return img + self.peso * ruido # Se aplica a la imagen después de multiplicar por el peso de cada canal.
    
    def get_peso(self):
        return self.peso

    def get_self(self):
        return self
    
```

## Normalización de instancia adaptable (AdaIN)
Para aumentar el control sobre la imagen, inyecta $ w $ (vector de ruido intermedio) varias veces en StyleGAN. Esto se hace transformándolo en un conjunto de parámetros de estilo e introduciendo el estilo a la imagen a través de AdaIN. Dada una imagen ($ x_i $) y el vector intermedio ($ w $), AdaIN toma la normalización de instancia de la imagen y la multiplica por la escala de estilo ($ y_s $) y agrega el sesgo de estilo ($ y_b $). Debe calcular la escala de estilo que se puede aprender y el sesgo mediante el uso de asignaciones lineales de $ w $.


 $ \text{AdaIN}(\boldsymbol{\mathrm{x}}_i, \boldsymbol{\mathrm{y}}) = \boldsymbol{\mathrm{y}}_{s,i} \frac{\boldsymbol{\mathrm{x}}_i - \mu(\boldsymbol{\mathrm{x}}_i)}{\sigma(\boldsymbol{\mathrm{x}}_i)} + \boldsymbol{\mathrm{y}}_{b,i} $


```python
class AdaIN(nn.Module):
    def __init__(self, inChan, w_dim): ## inChan: num canales de la imagen, w_dim: dimension del vector de ruido
        super().__init__()
        self.instance_norm = nn.InstanceNorm2d(inChan) # Normalizar la entrada por dimensión /Aplica la normalización de instancias sobre una entrada 4D
        #Mapeo w pesos de estilo (style weights) por canal. 
        self.tranformEscalaEstilo = nn.Linear(w_dim, inChan)
        self.tranformCambioEstilo = nn.Linear(w_dim, inChan)

    def forward(self, img, w): #Dada una imagen y un vector de ruido intermedio w, devuelve la imagen normalizada que ha sido escalada y cambiada por el estilo.
        imgNormalizada = self.instance_norm(img)
        EscalaStyle = self.tranformEscalaEstilo(w)[:, :, None, None]
        CambioStyle = self.tranformCambioEstilo(w)[:, :, None, None]

        imgTranformada = EscalaStyle * imgNormalizada + CambioStyle # Calculamos la imagen tranformada  
        return imgTranformada
    
    def get_style_scale_transform(self):
        return self.tranformEscalaEstilo
    

    def get_style_shift_transform(self):
        return self.tranformCambioEstilo
    
    def get_self(self):
        return self 
```

## Crecimiento progresivo en StyleGAN
Componente final para la Style GAN: permite crear imágenes de alta resolución duplicando gradualmente el tamaño de la imagen hasta el tamaño que deseamos.
Comenzará creando un bloque para el generador StyleGAN. Se compone de una capa de muestreo superior, una capa convolucional, inyección de ruido aleatorio, una capa AdaIN y una activación.

```python
class MicroStyleGAN_BloqueGenerador(nn.Module): 
    #tamIni: el tamaño de la imagen inicial
    def __init__(self, inChan, outChan, w_dim, kernTam, tamIni, usarUpsample=True):
        super().__init__()
        self.usar_upsample = usarUpsample

        if self.usar_upsample: #Upsample el tamaño inicial de forma bilinear
            self.upsample = nn.Upsample((tamIni), mode='bilinear')
        self.conv = nn.Conv2d(inChan, outChan, kernTam, padding=1) # Con padding mantenemos el tamaño de la imagen
        self.ruidoIny = InyeccionDeRuido(outChan)
        self.adain = AdaIN(outChan, w_dim)
        self.activacion = nn.LeakyReLU(0.2)


    def forward(self, x, w):#Dada una x y una w,calcula un bloque generador StyleGAN.
        #x: la entrada en el generador, mapa de características de la forma (n_samples, canales, ancho, alto)
        #w: vector de ruido
        if self.usar_upsample:
            x = self.upsample(x)
        x = self.conv(x)
        x = self.ruidoIny(x)
        x = self.activacion(x)
        x = self.adain(x, w)
        return x
    
    def get_self(self):
        return self;
```

Crecimiento progresivo:

StyleGAN comienza con un tensor constante de 4 x 4 (x 512 canales) que se somete a una iteración del generador sin sobremuestreo. La salida es algo de ruido que luego se puede transformar en una imagen borrosa de 4 x 4. Aquí es donde comienza el proceso de crecimiento progresivo. El ruido de 4 x 4 se puede pasar más a través de un bloque generador con muestreo superior para producir una salida de 8 x 8. Sin embargo, esto se hará gradualmente.

Simulará el crecimiento progresivo de una imagen de 8 x 8 a una imagen de 16 x 16. En lugar de simplemente pasarlo al bloque del generador con un muestreo superior, StyleGAN entrena gradualmente el generador al nuevo tamaño mezclando una imagen que solo fue muestreada. Al mezclar una imagen de 8 x 8 muestreada (que es de 16 x 16) con cada vez más de la salida del generador de 16 x 16, el generador es más estable a medida que se entrena progresivamente. Como tal, hará dos operaciones separadas con el ruido de 8 x 8:

1.   Pásalo al siguiente bloque generador para crear un ruido de salida, que luego transformará en una imagen.
2.   Transfórmalo en una imagen y luego amplíelo para que sea de 16 x 16.

Ahora tendrá dos imágenes que tienen el doble de resolución que el ruido de 8 x 8. Luego, usando un término alfa (𝛼), combina las imágenes de mayor resolución obtenidas de (1) y (2). Luego pasaría esto al discriminador y usaría la retroalimentación para actualizar los pesos de su generador. La clave aquí es que el término 𝛼 se incrementa gradualmente hasta que, finalmente, solo se usa la imagen de (1), el generador. Esa es su imagen final o puede continuar este proceso para hacer una imagen de 32 x 32 o 64 x 64, 128 x 128, etc.

Este micro modelo que implementará visualizará lo que genera el modelo en una etapa particular del entrenamiento, para un valor específico de $\alpha$. Sin embargo, para reiterar, en la práctica, StyleGAN eliminará gradualmente la imagen muestreada aumentando el parámetro $\alpha$ en muchos pasos de entrenamiento, haciendo este proceso repetidamente con valores alfa cada vez más grandes hasta que sea 1; en este punto, la imagen combinada está compuesta únicamente de la imagen del bloque generador. Este método de entrenamiento gradual del generador aumenta la estabilidad y fidelidad del modelo.


```python
class MicroStyleGANGenerator(nn.Module): #map_hidden_dim: la dimensión interna del mapeo, un escalar
    def __init__(self, 
                 z_dim, 
                 map_hidden_dim,
                 dimW,
                 inChan,
                 outChan, 
                 kernTam, 
                 hiddenChan):
        super().__init__()
        self.map = MapeoCapas(z_dim, map_hidden_dim, dimW)
        self.constanteIni = nn.Parameter(torch.randn(1, inChan, 4, 4))
        self.bloque0 = MicroStyleGAN_BloqueGenerador(inChan, hiddenChan, dimW, kernTam, 4, usarUpsample=False)
        self.bloque1 = MicroStyleGAN_BloqueGenerador(hiddenChan, hiddenChan, dimW, kernTam, 8)
        self.bloque2 = MicroStyleGAN_BloqueGenerador(hiddenChan, hiddenChan, dimW, kernTam, 16)
        # Para mapear desde el ruido de salida auna imagen: aprendemos con una conv con un kernel 1x1 
        ## por ejm 512 canales a 3 canales
        self.bloque1_a_img = nn.Conv2d(hiddenChan, outChan, kernel_size=1)
        self.bloque2_a_img = nn.Conv2d(hiddenChan, outChan, kernel_size=1)
        self.alpha = 0.2

    def upsampleAImagenSalida(self, imgPeq, imgGrande):
        return F.interpolate(imgPeq, size=imgGrande.shape[-2:], mode='bilinear')

    def forward(self, ruido, devolver_interm=False):
        x = self.constanteIni
        w = self.map(ruido)
        x = self.bloque0(x, w)
        x_pq = self.bloque1(x, w) # First generator run output
        x_imgPeq = self.bloque1_a_img(x_pq)
        x_grande = self.bloque2(x_pq, w) # Second generator run output 
        x_imgGrande = self.bloque2_a_img(x_grande)
        x_upsamplePeq = self.upsampleAImagenSalida(x_imgPeq, x_imgGrande) 
        # Interpolar entre la imagen muestreada y la imagen del generador usando alfa
        interpol = self.alpha * (x_imgGrande) + (1-self.alpha) * (x_upsamplePeq)

        if devolver_interm:
            return interpol, x_upsamplePeq, x_imgGrande
        return interpol
    
    def get_self(self):
        return self;
```

```python
class MicroStyleGANDiscrim(nn.Module): 
    #tamIni: el tamaño de la imagen inicial
    def __init__(self, inChan, outChan, w_dim, kernTam, tamIni, usarUpsample=True):
        super().__init__()
        self.from_rgbs = nn.ModuleList([
            SConv2d(3, 16, 1),
            SConv2d(3, 32, 1),
            SConv2d(3, 64, 1),
            SConv2d(3, 128, 1),
            SConv2d(3, 256, 1),
            SConv2d(3, 512, 1),
            SConv2d(3, 512, 1),
            SConv2d(3, 512, 1),
            SConv2d(3, 512, 1)
       ])
        self.convs  = nn.ModuleList([
            ConvBlock(16, 32, 3, 1),
            ConvBlock(32, 64, 3, 1),
            ConvBlock(64, 128, 3, 1),
            ConvBlock(128, 256, 3, 1),
            ConvBlock(256, 512, 3, 1),
            ConvBlock(512, 512, 3, 1),
            ConvBlock(512, 512, 3, 1),
            ConvBlock(512, 512, 3, 1),
            ConvBlock(513, 512, 3, 1, 4, 0)
        ])
        self.fc = SLinear(512, 1)
        
        self.numCapas = 9
        
    def forward(self, img, 
               steps=0, #cuantas capas se usa para entrenar
               alpha=-1):
        for i in range(step, -1, -1):
            indexCapa = self.numCapas - i - 1 
            
            #Primera capa: usamos from_rgb y convertirlo a numCanales 
            if i == step: 
                res = self.from_rgbs[indexCapa](img)
            
            # Antes de ultima capa, hacer minibatch stddev
            if i == 0:
                res_var = res.var(0, unbiased=False) + 1e-8 
                res_std = torch.sqrt(res_var)
                mean_std = res_std.mean().expand(result.size(0), 1, 4, 4)
                result = torch.cat([result, mean_std], 1)
            
            # Conv
            res = self.convs[indexCapa](res)
            
            # Capas intermedias
            if i > 0:
                # Downsample
                res = nn.functional.interpolate(result, scale_factor=0.5, mode='bilinear',align_corners=False)
                # Alpha , combianr resultados de diferentes capas
                if i == step and 0 <= alpha < 1:
                    res_sig = self.from_rgbs[indexCapa + 1](img)
                    res_sig = nn.functional.interpolate(res_sig, scale_factor=0.5,mode = 'bilinear', align_corners=False)
                
                    res = alpha * res + (1 - alpha) * res_sig

        res = res.squeeze(2).squeeze(2)
        res = self.fc(res)
        return res
    
    def get_self(self):
        return self;
```
